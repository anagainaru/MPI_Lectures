```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

# Performance for computing PI
- Running on Knight's Landing (Intel(R) Xeon Phi(TM) CPU 7230 @ 1.30GHz)
- space divided in 6.4 million slices

```python
pi_procs = [1<<i for i in range(1,7)]
pi_procs
```

```python
serial_run = 265.141814
pi_runs = [133.086422, 66.761714, 33.570689, 16.770134, 8.388822, 4.384464]
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.set(font_scale=2)
sns.set_style("white")
ax.plot(pi_procs, pi_runs, 'o-', lw=3, alpha=0.8)

ax.set_xlabel('Number of processes')
ax.set_ylabel('Execution time')

ax.set_xscale("log", basex=2)
#ax.set_yscale("log", basey=2)
plt.savefig("figures/pi_scalability.png")
plt.show()
```

```python
# plot speedup
fig, ax = plt.subplots(figsize=(12,8))
sns.set(font_scale=2)
sns.set_style("white")
ax.plot(pi_procs, [serial_run/i for i in pi_runs], 'o-', lw=3, alpha=0.8)
ax.plot(pi_procs, pi_procs, '--', lw=3, alpha=0.8)

ax.set_xlabel('Number of processes')
ax.set_ylabel('Speed-up')

ax.set_xscale("log", basex=2)
ax.set_yscale("log", basey=2)
plt.savefig("figures/pi_speed-up.png")
plt.show()
```
